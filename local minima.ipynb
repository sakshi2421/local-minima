```python
x = 2
lr=0.01
precision =0.0001
prevevous_step_size=1
max_iter=10000
iters=1
gf=lambda x:(x+3)**2
```

```python
import matplotlib.pyplot as plt
```

```python
gd=[]
```

```python
while precision<prevevous_step_size and iters<max_iter:
    prev=x
    x=x-lr*gf(prev)
    prevevous_step_size= abs(x-prev)
    iters+=1
    print ("iteration:" ,iters,"value",x)
    gd.append(x)
    
```

```python
print ("Local minima: ",x)
```

```python
plt.plot(gd)
plt.xlabel("Iteration")
plt.ylabel("Value of X")
plt.title("Gradient desent algorithm")
plt.show()
```
